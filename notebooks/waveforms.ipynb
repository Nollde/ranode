## Introduction

This notebook aims to show two different possibilities to generate the toy signals for the two detectors.

It should facilitate the discusion, which one is correct.

```python
import numpy as np
import scipy
from matplotlib import pyplot as plt
```

```python
amp = 100
period = 20
delta = 5
noise = 0
samples = 10000
```

```python
time_steps = np.linspace(0, 300, num=samples)
```

## Option 1
This is the way it was before, the detectors are not only shifted but the peak is also at another point of the sinus curve.

```python
signal_h = np.sin(2 * np.pi * (time_steps) / period) * scipy.stats.norm.pdf(time_steps, loc=150, scale=20)
signal_l = np.sin(2 * np.pi * (time_steps+delta) / period) * scipy.stats.norm.pdf(time_steps, loc=150+delta, scale=20)
plt.plot(time_steps, signal_h, label="H")
plt.plot(time_steps, signal_l, label="L")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Option 1")
plt.legend()
```

## Option 2
The measured curves are exactly that same but shifted in time.

```python
signal_h = np.sin(2 * np.pi * (time_steps) / period) * scipy.stats.norm.pdf(time_steps, loc=150, scale=20)
signal_l = np.sin(2 * np.pi * (time_steps+delta) / period) * scipy.stats.norm.pdf(time_steps+delta, loc=150, scale=20)
plt.plot(time_steps, signal_h, label="H")
plt.plot(time_steps, signal_l, label="L")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Option 2")
plt.legend()
```
